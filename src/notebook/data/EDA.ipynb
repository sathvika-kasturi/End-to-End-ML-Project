```python
### numerical features and categorical features
num_features=[feature for feature in df.columns if df[feature].dtypes != 'O']
cat_features=[feature for feature in df.columns if df[feature].dtypes=='O']
```

```python
num_features,cat_features
```

```python
## total score and average score
df['total_score']=df['math_score']+df['reading_score']+df['writing_score']
df['averag_score']=df['total_score']/3
```

```python
df['averag_score']
```

```python
df['total_score']
```

```python
df.head(3)
```

```python
## finding the number of students who got full marks
reading_full_score = (df['reading_score'] == 100).sum()
maths_full_score=(df['math_score']==100).sum()
writing_full_score=(df['writing_score']==100).sum()
print(f'number of students got full marks in reading_full_score:{reading_full_score}')
print(f'number of students got full marks in writing_full_score:{writing_full_score}')
print(f'number of students got full marks in maths_full_score:{maths_full_score}')


```

```python
# finding the number of students who got less than 50 marks
reading_full_score = (df['reading_score'] <=50).sum()
maths_full_score=(df['math_score']<=50).sum()
writing_full_score=(df['writing_score']<=50).sum()
print(f'number of students got less than 50  marks in reading_full_score:{reading_full_score}')
print(f'number of students got less than 50  marks in writing_full_score:{writing_full_score}')
print(f'number of students got less than 50 marks in maths_full_score:{maths_full_score}')
```

```python
## histogram and KDE
fig,axs=plt.subplots(1,2,figsize=(15,7))
plt.subplot(121)
sns.histplot(data=df,x='averag_score',bins=30,kde=True,color='r')
plt.subplot(122)
sns.histplot(data=df, x='averag_score', kde=True, hue='gender', ax=axs[1])
plt.show()
```
